```python
import numpy as np
import time
# Selection Sort algorithm
def sel_sort(arr):
  start = time.time()
  n = len(arr)
  for i in range(n):
    min = arr[i]
    min_idx = i
    for j in range(i+1,n):
      if(arr[j]<min):
        min = arr[j]
        min_idx = j
    temp = arr[i]
    arr[i] = arr[min_idx]
    arr[min_idx] = temp
  end = time.time()
  t = end - start
  return arr,t

arr = np.random.randint(100, size=24)
print(arr)
print(sel_sort(arr)[0])

times = []
for i in range(1024):
  arr = np.random.randint(100, size=24)
  times.append(sel_sort(arr)[1])
print(times)

times = np.array(times)

print('Mean time taken to sort the array = ',np.mean(times))
print('Standard deviation of all time taken to sort the arrays',np.std(times))


```

```python
'''
1. find min element in array and it's index
2. delete min element form array and append to ans
3. repeat till array is empty
'''

def sort2(arr):
  n = len(arr)
  ans = []
  for i in range(n):
    x = np.min(arr)
    idx = np.argmin(arr)
    arr = np.delete(arr, idx)
    ans.append(x)
  ans = np.array(ans)
  return ans


arr = np.random.randint(10, size=10)
print(arr)
print(sort2(arr))
```

```python
# Random walk problem
Nreal = 1200
avg_N = []
NS = []
for Nsteps in range(32,1000):
  disp_r = []
  for i in range(Nreal):
    steps = np.random.randint(2, size=Nsteps)
    steps[np.where(steps==0)] = -1
    d = np.sum(steps)
    disp_r.append(d)
  disp_r = np.array(disp_r)
  disp_r = np.abs(disp_r)
  avg = np.mean(disp_r)
  avg_N.append(avg)
  NS.append(Nsteps)
avg_N = np.array(avg_N)
NS = np.array(NS)

print(avg_N)

import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]=20,20
plt.plot(NS, np.abs(avg_N),)
y = np.power(NS*pow(2/np.pi,0.5),0.5)
plt.plot(NS,y,'r')
```
