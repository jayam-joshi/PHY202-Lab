```python
## Q1
import numpy as np
import random
import matplotlib.pyplot as plt
import pylab
N =1000
xx = np.zeros(N)
yy = np.zeros(N)
for i in range(1,N):
  step = random.choice((1,2,3,4))
  if(step == 1):
    xx[i] = xx[i-1] + 1
    yy[i] = yy[i-1]
  elif(step == 2):
    xx[i] = xx[i-1] - 1
    yy[i] = yy[i-1]
  elif(step==3):
    yy[i] = yy[i-1] + 1
    xx[i] = xx[i-1]
  elif(step==4):
    yy[i] = yy[i-1] - 1
    xx[i] = xx[i-1]

plt.figure(1)
pylab.plot(xx,yy)

Nreal = 1000
R2_N = []
for Nsteps in range(10,100):
  r2_n = []
  for j in range(Nreal):
    x = 0
    y = 0
    for i in range(Nsteps):
      step = random.choice((1,2,3,4))
      if(step == 1):
        x = x + 1
      elif(step == 2):
        x = x - 1
      elif(step == 3):
        y = y + 1
      elif(step == 4):
        y = y - 1
    r2 = x**2 + y**2
    r2_n.append(r2)
  r2_n = np.array(r2_n)
  R2_N.append(np.mean(r2_n))

R2_N = np.array(R2_N)
NN = np.arange(10,100)

plt.figure(2)
plt.xlabel('$log(N)$')
plt.ylabel('$log(<R^2(N)>)$')
x = np.log(NN)
y = np.log(R2_N)
plt.plot(x, y, color='black')

slope, intercept = np.polyfit(x, y, 1)
line = slope*x + intercept
nu = slope
print('Exponent = ',nu)      ## Exponent for 2D    appprox = 1

plt.plot(x, line, color = 'red')
plt.text(2.5,4.5, 'Exponent = %f'%nu)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import random
import matplotlib.pyplot as plt

N =1000
xx = np.zeros(N)
yy = np.zeros(N)
zz = np.zeros(N)
for i in range(1,N):
  step = random.choice((1,2,3,4,5,6))
  if(step == 1):
    xx[i] = xx[i-1] + 1
    yy[i] = yy[i-1]
    zz[i] = zz[i-1]
  elif(step == 2):
    xx[i] = xx[i-1] - 1
    yy[i] = yy[i-1]
    zz[i] = zz[i-1]
  elif(step==3):
    yy[i] = yy[i-1] + 1
    xx[i] = xx[i-1]
    zz[i] = zz[i-1]
  elif(step==4):
    yy[i] = yy[i-1] - 1
    xx[i] = xx[i-1]
    zz[i] = zz[i-1]
  elif(step==5):
    zz[i] = zz[i-1] + 1
    xx[i] = xx[i-1]
    yy[i] = yy[i-1]
  elif(step==6):
    zz[i] = zz[i-1] - 1
    xx[i] = xx[i-1]
    yy[i] = yy[i-1]

plt.figure(1)
ax = plt.subplot(1,1,1, projection='3d')
ax.plot(xx, yy, zz,alpha=0.9) 


Nreal = 10000
R2_N = []
for Nsteps in range(10,100):
  r2_n = []
  for j in range(Nreal):
    x = 0
    y = 0
    z = 0
    for i in range(Nsteps):
      step = random.choice((1,2,3,4,5,6))
      if(step == 1):
        x = x + 1
      elif(step == 2):
        x = x - 1
      elif(step == 3):
        y = y + 1
      elif(step == 4):
        y = y - 1
      elif(step == 5):
        z = z + 1
      elif(step == 6):
        z = z - 1

    r2 = x**2 + y**2 + z**2
    r2_n.append(r2)
  r2_n = np.array(r2_n)
  R2_N.append(np.mean(r2_n))

R2_N = np.array(R2_N)
NN = np.arange(10,100)

x = np.log(NN)
y = np.log(R2_N)


plt.figure(2)
plt.xlabel('$log(N)$')
plt.ylabel('$log(<R^2(N)>)$')
plt.plot(x, y, color='black')

slope, intercept = np.polyfit(x, y, 1)
line = slope*x + intercept
nu = slope
print('Exponent = ',nu)      ## Exponent for 2D    appprox = 1

plt.plot(x, line, color = 'red')
plt.text(2.5,4.5, 'Exponent = %f'%nu)


```
